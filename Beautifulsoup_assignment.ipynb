### 1. Write a python program to display all the header tags from wikipedia.org and make data frame.


```python
page1 = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page1
```

```python
soup1 = BeautifulSoup(page1.content)
soup1
```

```python
all_headers = []

for i in soup1.find_all("span", class_='mw-headline'):
    all_headers.append(i.text)
    
all_headers
```

```python
# making data frame
df_headers = pd.DataFrame({'All Headers':all_headers})
df_headers
```

### 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame

```python
page5 = requests.get("https://www.cnbc.com/world/?region=world")
page5
```

```python
soup5 = BeautifulSoup(page5.content)
```

```python
headline=[]
for i in soup5.find_all("a", class_=["LatestNews-headline"]):
    headline.append(i.text)
    
headline
```

```python
time=[]
for i in soup5.find_all("time", class_=["LatestNews-timestamp"]):
    time.append(i.text)
    
time
```

```python
#for link in soup5.find_all("a", class_="LatestNews-headline"):
    #print(link.get("href"))
    
link =[]
for i in soup5.find_all("a",class_="LatestNews-headline"):
    link.append(i["href"])
link
```

```python
news_details = pd.DataFrame({"Headline":headline,"Time":time,"Links":link})
news_details
```

### 7) Write a python program to scrape mentioned details from dineout.co.inand make data frame

```python
page7 = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page7
```

```python
soup7 = BeautifulSoup(page7.content)
```

```python
name = []
for i in soup7.find_all("a", class_=["restnt-name ellipsis"]):
    name.append(i.text)
    
name
```

```python
cuisine=[]
for i in soup7.find_all("span",class_="double-line-ellipsis"):
    cuisine.append(i.text)
cuisine
```

```python
location=[]
for i in soup7.find_all("div",class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
ratings=[]
for i in soup7.find_all("div",class_="restnt-rating rating-4"):
    ratings.append(i.text)
ratings
```

```python
img_url =[]
for i in soup7.find_all("img",class_="no-img"):
    img_url.append(i["data-src"])
img_url
```

```python
df7=pd.DataFrame({"Restaurant Name":name,"Cuisine":cuisine,"Location":location,"Ratings":ratings,"Image URL":img_url})
df7
```
